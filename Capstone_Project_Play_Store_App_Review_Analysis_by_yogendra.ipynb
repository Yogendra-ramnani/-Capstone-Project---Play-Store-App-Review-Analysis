**Rating column NaN values handling**

As we can see Rating column contain 1470 NaN values. it's a large number of data so we can't drop all the rows if we do that than we will loose large amount of data. Hence we are going to replace NaN values with the mean or median of the Rating column present values.


That help us in further analysis and mining. 

```python
# Rows which are containing NaN values in the Rating column
ps_df[ps_df['Rating'].isnull()]
```

```python
# Boxplot
from numpy.ma.extras import median

plot_figure = plt.figure(figsize =(10,5))
sns.boxplot(data = ps_df['Rating'] ,orient="h", x= ps_df['Rating'])

median = ps_df['Rating'].median()

plt.title ("Boxplot of Rating values")
```

**Displot**- it's help us to find the mean and median values according to values alredy present in Rating column.

```python
#Displot [Plot before the handle of NaN values]
sns.set(rc={'figure.figsize':(12,7)})
sns.distplot(ps_df['Rating'])
```

From the above Boxplot and Displot graph, we can come to the conclusion that most of the apps in the google play store are rated between 3.2 to 4.8. and also

1.Mean = 4.19 (excluding Nan values)

2.Median = 4.30 (excluding NaN values)

we alose note that the graph is left skewed and is not catagorical, it is not recomamended to use mean for replacing the NaN value. 
Hence we will impute the NaN values in the Rating column with its median.

```python
# Cross checking the value of median in the 'Rating' column
ps_df['Rating'].median()
```

```python
# Replacing the NaN values in the 'Rating' colunm with its median value which is 4.3
impute_median = ps_df['Rating'].median()
ps_df['Rating'] = ps_df['Rating'].fillna(impute_median)
```

```python
# Plotting histogram
ps_df ['Rating'].hist()
```

After treatinge and replacing the NaN values we observe there is no major devation in orignal plot.

# **Changing in columns Dtype**

1. Change and clean the string values present into different columns.
2. Converting 'Review','Price', 'Installs' and 'Size' columns into valid numeric columns.



```python
ps_df.info()
```

```python
# Reviews column
## Printing unique values of Reviews column
ps_df[('Reviews')].unique()
```

```python
# Convering 'Reviews" column inot valid numeric column
ps_df[('Reviews')] = pd.to_numeric(ps_df['Reviews'])
```

```python
# Price column
## Printing unique values of price column
ps_df[('Price')].unique()
```

```python
# Removing undefine symbols['$' and ',']
symb_to_remove = ['$', ',']
colum_name = ['Price']

for symb in symb_to_remove:
  ps_df['Price'] = ps_df['Price'].astype(str).str.replace(symb, '', regex=True)
```

```python
# Convering 'Price' column inot valid numeric column
ps_df['Price'] = pd.to_numeric(ps_df['Price'])
```

```python
ps_df['Price'].unique()
```

```python
# Installs column
## Printing unique values of Installs column
ps_df[('Installs')].unique()
```

```python
# Removing undefine symbols['+' and ',']
symb_to_remove = ['+', ',']
colum_name = ['Installs']

for symb in symb_to_remove:
  ps_df['Installs'] = ps_df['Installs'].astype(str).str.replace(symb, '', regex=True)
```

```python
# Convering 'Installs' column inot valid numeric column
ps_df['Installs'] = pd.to_numeric(ps_df['Installs'])
```
