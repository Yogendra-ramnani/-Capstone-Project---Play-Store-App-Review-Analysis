**1 - "Review" Column** - "Review" column does not containg any type of symbol so we can easily change "Review" column data into numeric type data.

```python
# Printing unique values of Reviews column

ps_df[('Reviews')].unique()
```

```python
# Convering 'Reviews" column inot valid numeric column

ps_df[('Reviews')] = pd.to_numeric(ps_df['Reviews'])
```

**2 - "Price" Column** - "Price" column containg different type of symbols E.g. '$' and ',' so we are going to remove those symobls and change "Price" column data into numeric type data.

```python
# Printing unique values of price column

ps_df[('Price')].unique()
```

```python
# Removing undefine symbols['$' and ',']

symb_to_remove = ['$', ',']
colum_name = ['Price']

for symb in symb_to_remove:
  ps_df['Price'] = ps_df['Price'].astype(str).str.replace(symb, '', regex=True)
```

```python
# Convering 'Price' column inot valid numeric column

ps_df['Price'] = pd.to_numeric(ps_df['Price'])
```

```python
ps_df['Price'].unique()
```

**3 - "Installs" Column** - "Installs" column containg different type of symbols E.g. '+' and ',' so we are going to remove those symobls and change "Installs" column data into numeric type data.

```python
# Installs column
## Printing unique values of Installs column

ps_df[('Installs')].unique()
```

```python
# Removing undefine symbols['+' and ',']

symb_to_remove = ['+', ',']
colum_name = ['Installs']

for symb in symb_to_remove:
  ps_df['Installs'] = ps_df['Installs'].astype(str).str.replace(symb, '', regex=True)
```

```python
# Convering 'Installs' column inot valid numeric column

ps_df['Installs'] = pd.to_numeric(ps_df['Installs'])
```

```python
ps_df['Installs'].unique()
```
